Непрерывная случайная величина X распределена нормально и задана плотностью распределения f(x) = (1 / (4 * sqrt(2pi))) * exp((-(x+2)**2) / 32)

Найдите:

а). M(X)

б). D(X)

в). std(X) (среднее квадратичное отклонение)

```python
print(f'M(X) = {-2}\n' # (x-u) -> (x+2) => u = M(X) = -2
      f'D(X) = {16}\n' # 2 * o ** 2 -> 32 => D(X) = 0 ** 2 = 16
      f'std(X) = {4}') # o * sqrt(2pi) -> 4 * sqrt(2pi) => o = sdx(X) = 4
```

Рост взрослого населения города X имеет нормальное распределение. Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

а). больше 182 см

б). больше 190 см

в). от 166 см до 190 см

г). от 166 см до 182 см

д). от 158 см до 190 см

е). не выше 150 см или не ниже 190 см

ё). не выше 150 см или не ниже 198 см

ж). ниже 166 см.

```python
from scipy import stats
M, std = 174, 8
print(f'Вероятность, что взрослый человек имеет рост больше 182 см = {round((1 - stats.norm.cdf(182, M, std)), 3)}\n'
      f'Вероятность, что взрослый человек имеет рост больше 190 см = {round(1 - stats.norm.cdf(190, M, std), 3)}\n'
      f'Вероятность, что взрослый человек имеет рост от 166 см до 190 см = {(0.954 - 0.68) / 2 + 0.68}\n'
      f'Вероятность, что взрослый человек имеет рост от 166 см до 182 см = {0.68}\n' # отклонение по 1 сигмы слева и справа
      f'Вероятность, что взрослый человек имеет рост от 158 см до 190 см = {0.954}\n' # отклонение по 2 сигмы слева и справа
      f'Вероятность, что взрослый человек имеет рост не выше 150 см или не ниже 190 см = {round(1 - ((0.9972 - 0.954) / 2 + 0.954), 3)}\n'
      f'Вероятность, что взрослый человек имеет рост не выше 150 см или не ниже 198 см = {round(1 - 0.9972, 3)}\n' # отклонение по 3 сигмы слева и справа
      f'Вероятность, что взрослый человек имеет рост ниже 166 см = {round(stats.norm.cdf(166, M, std), 3)}')
```

На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?
