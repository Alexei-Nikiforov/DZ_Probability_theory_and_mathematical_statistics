Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16.
Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256.

```python
from scipy import stats
x = 80
n = 256
s = 16

# Стандартная ошибка
se = s / n ** 0.5

# Значение z-критерия для 0.95% доверительной вероятности
z_tabl = stats.norm.ppf(0.975)

# Границы доверительного интервала
print(x - se * z_tabl, x + se * z_tabl)
```

В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные:

6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1

Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

```python
import numpy as np
from scipy import stats
arr = np.array([6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1])
n = 10

# Среднее арифметическое по выборке
x = np.mean(arr)

# Среднее квадратическое отклонение по выборке
s = np.std(arr, ddof = 1)

# Стандартная ошибка
se = s / n ** 0.5

# Значение t-критерия для 0.95% доверительной вероятности
t_95 = stats.t.ppf(0.975, n - 1)

# Границы доверительного интервала
print(x - se * t_95, x + se * t_95)
```

Используя эти данные построить 95% доверительный интервал для разности среднего роста родителей и детей.

```python
import numpy as np
from scipy import stats
arr1 = np.array([175, 167, 154, 174, 178, 148, 160, 167, 169, 170])
arr2 = np.array([178, 165, 165, 173, 168, 155, 160, 164, 178, 175])

# Размер выборки
n1 = len(arr1)
n2 = len(arr2)

# Среднее арифметическое по выборке
x1 = np.mean(arr1)
x2 = np.mean(arr2)

# Среднее квадратическое отклонение по выборке
s1 = np.std(arr1, ddof = 1)
s2 = np.std(arr2, ddof = 1)

# Разность средних
delta = abs(x1 - x2)

# Стандартная ошибка разности средних
se = (s1 ** 2 / n1 + s2 ** 2 / n2) ** 0.5

# Значение t-критерия для 0.95% доверительной вероятности
t_95 = stats.t.ppf(0.975, n1 + n2 - 2)

# Границы доверительного интервала
print(delta - se * t_95, delta + se * t_95)
```
