1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
import numpy as np

def arithmetic_mean (array):
    return sum(array) / len(array)

def mean_square_deviation (array):
    square_deviation = (array - arithmetic_mean(array)) ** 2
    return (sum(square_deviation) / (len(square_deviation) - 1)) ** (1/2)

def biased_variance (array):
    square_deviation = (array - arithmetic_mean(array)) ** 2
    return (sum(square_deviation) / len(square_deviation))

def unbiased_variance (array):
    square_deviation = (array - arithmetic_mean(array)) ** 2
    return (sum(square_deviation) / (len(square_deviation) - 1))

a = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])

print(f'Среднее арифметическое = {arithmetic_mean(a)}\n'
      f'Среднее квадратичное отклонение = {mean_square_deviation(a)}\n'
      f'Смещенная оценка дисперсии = {biased_variance(a)}\n'
      f'Немещенная оценка дисперсии = {unbiased_variance(a)}\n')
```

```python
import numpy as np
a = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
print(f'Среднее арифметическое = {np.mean(a)}\n'
      f'Среднее квадратичное отклонение = {np.std(a, ddof=1)}\n'
      f'Смещенная оценка дисперсии = {np.var(a, ddof=0)}\n'
      f'Немещенная оценка дисперсии = {np.var(a, ddof=1)}\n')
```

2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
from math import factorial
def combinations(n, k):
    return factorial(n) / (factorial(n - k) * factorial(k))

# 1 случай: из 1 ящика - 0 белых, из 2 ящика - 3 белых
p_1 = (combinations(5,0) * combinations(3,2) / combinations(8,2))
p_2 = (combinations(5,3) * combinations(7,1) / combinations(12,4))
P1 = p_1 * p_2 # вероятность 1 случая

# 2 случай: из 1 ящика - 1 белый, из 2 ящика - 2 белых
p_1 = (combinations(5,1) * combinations(3,1) / combinations(8,2))
p_2 = (combinations(5,2) * combinations(7,2) / combinations(12,4))
P2 = p_1 * p_2 # вероятность 2 случая

# 3 случай: из 1 ящика - 2 белый, из 2 ящика - 1 белый
p_1 = (combinations(5,2) * combinations(3,0) / combinations(8,2))
p_2 = (combinations(5,1) * combinations(7,3) / combinations(12,4))
P3 = p_1 * p_2 # вероятность 3 случая

P = P1 + P2 + P3 # общая вероятность
print(f'Вероятность того, что все 3 мяча белые = {P}')
```

3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
# A - попадание, В - спортсмен
P_BA = 1/3
P_A1, P_A2, P_A3 = 0.9, 0.8, 0.6 

# Общая вероятность попадания всех 3 спортсменов
P_B = P_BA * P_A1 + P_BA * P_A2 + P_BA * P_A3

# Используя формулу Байса найдем вероятность попадания каждого спортсмена
P_AB_1 = P_BA * P_A1 / P_B
P_AB_2 = P_BA * P_A2 / P_B
P_AB_3 = P_BA * P_A3 / P_B

print(f'Вероятность, что выстрел сделал 1 спортсмен = {P_AB_1}\n'
      f'Вероятность, что выстрел сделал 2 спортсмен = {P_AB_2}\n'
      f'Вероятность, что выстрел сделал 3 спортсмен = {P_AB_3}')
```

4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
# A - сданная сессия, В - факультет
P_BA_fac_A, P_BA_fac_B, P_BA_fac_C = 1/4, 1/4, 1/2
P_A_fac_A, P_A_fac_B, P_A_fac_C = 0.8, 0.7, 0.9 

# Общая вероятность сдачи первой сессии на всех 3 факультетах
P_B_fac = P_BA_fac_A * P_A_fac_A + P_BA_fac_B * P_A_fac_B + P_BA_fac_C * P_A_fac_C

# Используя формулу Байса найдем вероятность сдачи первой сессии на каждом факультете
P_AB_fac_A = P_BA_fac_A * P_A_fac_A / P_B_fac
P_AB_fac_B = P_BA_fac_B * P_A_fac_B / P_B_fac
P_AB_fac_C = P_BA_fac_C * P_A_fac_C / P_B_fac

print(f'Вероятность, что студент сдал первую ссесию на факультете A = {P_AB_fac_A}\n'
      f'Вероятность, что студент сдал первую ссесию на факультете B = {P_AB_fac_B}\n'
      f'Вероятность, что студент сдал первую ссесию на факультете C = {P_AB_fac_C}')
```

5. Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25.

Какова вероятность того, что в первый месяц выйдут из строя: а). все детали?

```python
p_A, p_B, p_C = 0.1, 0.2, 0.25
P = p_A * p_B * p_C
print(f'Вероятность, что в первый месяц выйдут из строя все детали = {P}')
```

Какова вероятность того, что в первый месяц выйдут из строя: б). только две детали?

```python
p_A, p_B, p_C = 0.1, 0.2, 0.25

# 1 случай - 1 и 2 детали выйдут из строя, 3 деталь продолжит работать
p_1 = p_A * p_B * (1 - p_C)

# 2 случай - 1 и 3 детали выйдут из строя, 2 деталь продолжит работать
p_2 = p_A * (1 - p_B) * p_C

# 3 случай - 2 и 3 детали выйдут из строя, 1 деталь продолжит работать
p_3 = (1 - p_A) * p_B * p_C

# Посчитаем общую вероятность
P = p_1 + p_2 + p_3
print(f'Вероятность, что в первый месяц выйдут из строя только две детали = {P}')
```

Какова вероятность того, что в первый месяц выйдут из строя: в). хотя бы одна деталь?

```python
p_A, p_B, p_C = 0.1, 0.2, 0.25

# Вероятность, что устройство будет работать
Q = (1 - p_A) * (1 - p_B) * (1 - p_C)

# Вероятность, что выйдет из строя хотя бы одна деталь
P = 1 - Q
print(f'Вероятность, что в первый месяц выйдет из строя хотя бы одна деталь = {P}')
```

Какова вероятность того, что в первый месяц выйдут из строя: г). от одной до двух деталей?

```python
p_A, p_B, p_C = 0.1, 0.2, 0.25

# 1 случай - 1 деталь выйдет из строя, 2 и 3 детали продолжат работать
p_4 = p_A * (1 - p_B) * (1 - p_C)

# 2 случай - 2 деталь выйдет из строя, 1 и 3 детали продолжат работать
p_5 = (1 - p_A) * p_B * (1 - p_C)

# 3 случай - 3 деталь выйдет из строя, 1 и 2 детали продолжат работать
p_6 = (1 - p_A) * (1 - p_B) * p_C

# 4 случай - 1 и 2 детали выйдут из строя, 3 деталь продолжит работать
p_4 = p_A * p_B * (1 - p_C)

# 5 случай - 1 и 3 детали выйдут из строя, 2 деталь продолжит работать
p_5 = p_A * (1 - p_B) * p_C

# 6 случай - 2 и 3 детали выйдут из строя, 1 деталь продолжит работать
p_6 = (1 - p_A) * p_B * p_C

# Посчитаем общую вероятность
P = p_1 + p_2 + p_3 + p_4 + p_5 + p_6
print(f'Вероятность, что в первый месяц выйдут из строя от одной до двух деталей = {P}')
```
